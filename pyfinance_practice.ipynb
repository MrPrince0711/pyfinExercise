Plotly is interactive thus a better option as it can be used in dashboards

```python
import plotly.offline as pyo
import plotly.graph_objects as gp 
from plotly.subplots import make_subplots 
pyo.init_notebook_mode(connected = True)
pd.options.plotting.backend = 'plotly'
```

```python
Close.plot() 
```

```python
df2 = pdr.get_data_yahoo('AGL.JO', start, end)
df2.head()
```

```python
df2['MA50'] = df2['Close'].rolling(window = 50, min_periods = 0).mean()
df2['MA200'] = df2['Close'].rolling(window = 200, min_periods = 0).mean()
```

```python
fig = make_subplots(rows = 2, cols = 1, shared_xaxes = True, vertical_spacing = 0.1, subplot_titles= ('AGL', 'Volume'), row_width = [0.2, 0.7])
```

```python
fig.add_trace(gp.Candlestick(x = df2.index, open = df2['Open'], high = df2['High'], low = df2['Low'], close = df2['Close'], name = 'highOpen_lowClose'), row = 1, col = 1)
```

```python
fig.add_trace(gp.Scatter(x = df2.index, y = df2['MA50'], marker_color = 'grey', name = 'MA50'), row = 1, col = 1)
fig.add_trace(gp.Scatter(x = df2.index, y = df2['MA200'], marker_color = 'blue', name = 'MA200'), row = 1, col = 1)
```

```python
fig.add_trace(gp.Bar(x = df2.index, y = df2['Volume'], marker_color = 'red', showlegend = False,), row = 2, col = 1)
```

```python
fig.update(layout_xaxis_rangeslider_visible = False)
```
